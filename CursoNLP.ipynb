# Clasificaci√≥n de disparo cero
Comenzaremos abordando una tarea m√°s desafiante en la que debemos clasificar textos que no han sido etiquetados. Este es un escenario com√∫n en proyectos del mundo real porque anotar texto suele llevar mucho tiempo y requiere experiencia en el dominio. Para este caso de uso, la zero-shot-classificationsecuencia de comandos es muy poderosa: le permite especificar qu√© etiquetas usar para la clasificaci√≥n, de modo que no tenga que depender de las etiquetas del modelo entrenado previamente. Ya ha visto c√≥mo el modelo puede clasificar una oraci√≥n como positiva o negativa usando esas dos etiquetas, pero tambi√©n puede clasificar el texto usando cualquier otro conjunto de etiquetas que desee.

```python




from transformers import pipeline

classifier1 = pipeline("zero-shot-classification")
classifier1(
    sequences="Jane bought me these books,\nJane bought a book for me,\nShe dropped a line to him. Thank you,\nShe sleeps,\nI sleep a lot,\nI was born in Madrid,\nthe cat was chased by the dog,\nI was born in Madrid during 1995,\nOut of all this , something good will come,\nSusan left after the rehearsal. She did it well,\nShe sleeps during the morning, but she sleeps.",
    candidate_labels=["education", "politics", "business"],
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

Esta secuencia se denomina zero-shot porque no es necesario ajustar el modelo en los datos para utilizarla. Puede devolver directamente puntuaciones de probabilidad para cualquier lista de etiquetas que desee.

# **Reconocimiento de entidades con nombre**
El reconocimiento de entidades con nombre (NER) es una tarea en la que el modelo debe encontrar qu√© partes del texto de entrada corresponden a entidades como personas, ubicaciones u organizaciones. Veamos un ejemplo:

```python
from transformers import pipeline

ner1 = pipeline("ner", grouped_entities=True)
ner1( "Jane bought me these books,\nJane bought a book for me,\nShe dropped a line to him. Thank you,\nShe sleeps,\nI sleep a lot,\nI was born in Madrid,\nthe cat was chased by the dog,\nI was born in Madrid during 1995,\nOut of all this , something good will come,\nSusan left after the rehearsal. She did it well,\nShe sleeps during the morning, but she sleeps.")
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")
```

Aqu√≠ el modelo identific√≥ correctamente que Sylvain es una persona (PER), Hugging Face una organizaci√≥n (ORG) y Brooklyn una ubicaci√≥n (LOC).

Pasamos la opci√≥n grouped_entities=True en la funci√≥n de creaci√≥n de la canalizaci√≥n para indicarle a la canalizaci√≥n que reagrupara las partes de la oraci√≥n que corresponden a la misma entidad: aqu√≠ el modelo agrup√≥ correctamente ‚ÄúHugging‚Äù y ‚ÄúFace‚Äù como una sola organizaci√≥n, aunque el nombre consta de varias palabras. De hecho, como veremos en el pr√≥ximo cap√≠tulo, el preprocesamiento incluso divide algunas palabras en partes m√°s peque√±as. Por ejemplo, Sylvain se divide en cuatro partes: S, ##yl, ##va y ##in. En el paso de posprocesamiento, la canalizaci√≥n reagrup√≥ esas partes con √©xito.

# **Capas de atenci√≥n**
Una caracter√≠stica clave de los modelos Transformer es que est√°n construidos con capas especiales llamadas capas de atenci√≥n. De hecho, el t√≠tulo del art√≠culo que presenta la arquitectura Transformer era ‚Äú¬°Atenci√≥n es todo lo que necesitas!‚Äù. Exploraremos los detalles de las capas de atenci√≥n m√°s adelante en el curso; por ahora, todo lo que necesitas saber es que esta capa le indicar√° al modelo que preste atenci√≥n espec√≠fica a ciertas palabras en la oraci√≥n que le pasaste (e ignore m√°s o menos las otras) cuando trabaje con la representaci√≥n de cada palabra.

Para poner esto en contexto, considere la tarea de traducir un texto del ingl√©s al franc√©s. Dado el dato de entrada ‚ÄúTe gusta este curso‚Äù, un modelo de traducci√≥n tambi√©n deber√° prestar atenci√≥n a la palabra adyacente ‚ÄúYou‚Äù para obtener la traducci√≥n adecuada de la palabra ‚Äúlike‚Äù, porque en franc√©s el verbo ‚Äúlike‚Äù se conjuga de manera diferente seg√∫n el sujeto. El resto de la oraci√≥n, sin embargo, no es √∫til para la traducci√≥n de esa palabra. En la misma l√≠nea, al traducir ‚Äúthis‚Äù, el modelo tambi√©n deber√° prestar atenci√≥n a la palabra ‚Äúcourse‚Äù, porque ‚Äúthis‚Äù se traduce de manera diferente seg√∫n si el sustantivo asociado es masculino o femenino. Nuevamente, las otras palabras en la oraci√≥n no importar√°n para la traducci√≥n de ‚Äúcurso‚Äù. Con oraciones m√°s complejas (y reglas gramaticales m√°s complejas), el modelo necesitar√≠a prestar especial atenci√≥n a las palabras que podr√≠an aparecer m√°s lejos en la oraci√≥n para traducir correctamente cada palabra.

El mismo concepto se aplica a cualquier tarea asociada con el lenguaje natural: una palabra por s√≠ misma tiene un significado, pero ese significado se ve profundamente afectado por el contexto, que puede ser cualquier otra palabra (o palabras) antes o despu√©s de la palabra que se est√° estudiando.

Ahora que tiene una idea de qu√© son las capas de atenci√≥n, echemos un vistazo m√°s de cerca a la arquitectura de Transformer.


# **La arquitectura original**
La arquitectura Transformer fue dise√±ada originalmente para la traducci√≥n. Durante el entrenamiento, el codificador recibe entradas (oraciones) en un idioma determinado, mientras que el decodificador recibe las mismas oraciones en el idioma de destino deseado. En el codificador, las capas de atenci√≥n pueden utilizar todas las palabras de una oraci√≥n (ya que, como acabamos de ver, la traducci√≥n de una palabra dada puede depender tanto de lo que est√° antes como despu√©s de ella en la oraci√≥n). El decodificador, sin embargo, trabaja de forma secuencial y solo puede prestar atenci√≥n a las palabras de la oraci√≥n que ya ha traducido (es decir, solo a las palabras anteriores a la palabra que se est√° generando actualmente). Por ejemplo, cuando hemos predicho las tres primeras palabras del destino traducido, se las damos al decodificador, que luego utiliza todas las entradas del codificador para intentar predecir la cuarta palabra.

Para acelerar el proceso de entrenamiento (cuando el modelo tiene acceso a las oraciones objetivo), el decodificador recibe todo el objetivo, pero no se le permite utilizar palabras futuras (si tuviera acceso a la palabra en la posici√≥n 2 al intentar predecir la palabra en la posici√≥n 2, el problema no ser√≠a muy dif√≠cil). Por ejemplo, al intentar predecir la cuarta palabra, la capa de atenci√≥n solo tendr√° acceso a las palabras en las posiciones 1 a 3.

La arquitectura original de Transformer se ve√≠a as√≠, con el codificador a la izquierda y el decodificador a la derecha:

Tenga en cuenta que la primera capa de atenci√≥n en un bloque de decodificador presta atenci√≥n a todas las entradas (pasadas) al decodificador, pero la segunda capa de atenci√≥n utiliza la salida del codificador. Por lo tanto, puede acceder a toda la oraci√≥n de entrada para predecir mejor la palabra actual. Esto es muy √∫til ya que diferentes idiomas pueden tener reglas gramaticales que colocan las palabras en diferentes √≥rdenes, o alg√∫n contexto proporcionado m√°s adelante en la oraci√≥n puede ser √∫til para determinar la mejor traducci√≥n de una palabra dada.

# Preprocesamiento con un tokenizador
Al igual que otras redes neuronales, los modelos Transformer no pueden procesar texto sin formato directamente, por lo que el primer paso de nuestro proceso es convertir las entradas de texto en n√∫meros que el modelo pueda interpretar. Para ello, utilizamos un tokenizador, que ser√° responsable de:

Dividir la entrada en palabras, subpalabras o s√≠mbolos (como la puntuaci√≥n) que se denominan tokens
Asignar cada token a un n√∫mero entero
A√±adir entradas adicionales que puedan resultar √∫tiles para el modelo
Todo este preprocesamiento debe realizarse exactamente de la misma manera que cuando se entren√≥ previamente el modelo, por lo que primero debemos descargar esa informaci√≥n del Model Hub. Para ello, utilizamos la clase AutoTokenizer y su m√©todo from_pretrained(). Utilizando el nombre del punto de control de nuestro modelo, obtendr√° autom√°ticamente los datos asociados con el tokenizador del modelo y los almacenar√° en cach√© (de modo que solo se descargue la primera vez que ejecute el c√≥digo a continuaci√≥n).

Dado que el punto de control predeterminado del proceso de an√°lisis de sentimientos es distilbert-base-uncased-finetuned-sst-2-english (puede ver su tarjeta de modelo aqu√≠), ejecutamos lo siguiente:


```python
from transformers import AutoTokenizer

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

Una vez que tenemos el tokenizador, podemos pasarle nuestras oraciones directamente y obtendremos un diccionario listo para alimentar a nuestro modelo. Lo √∫nico que queda por hacer es convertir la lista de ID de entrada en tensores.

Puede usar ü§ó Transformers sin tener que preocuparse por qu√© marco de ML se usa como backend; puede ser PyTorch o TensorFlow, o Flax para algunos modelos. Sin embargo, los modelos Transformer solo aceptan tensores como entrada. Si es la primera vez que escucha sobre tensores, puede pensar en ellos como matrices NumPy. Una matriz NumPy puede ser un escalar (0D), un vector (1D), una matriz (2D) o tener m√°s dimensiones. Es efectivamente un tensor; los tensores de otros marcos de ML se comportan de manera similar y, por lo general, son tan simples de instanciar como las matrices NumPy.

Para especificar el tipo de tensores que queremos recuperar (PyTorch, TensorFlow o NumPy simple), usamos el argumento

```python
raw_inputs1 = [  """
    Jane bought me these books.
Jane bought a book for me.
She dropped a line to him. Thank you.
She sleeps.
I sleep a lot.
I was born in Madrid.
the cat was chased by the dog.
I was born in Madrid during 1995.
Out of all this , something good will come.
Susan left after the rehearsal. She did it well.
She sleeps during the morning, but she sleeps.
"""]

inputs1 = tokenizer(raw_inputs1, padding=True, truncation=True, return_tensors="pt")
print(inputs1)
```

padding: Las entradas por lotes suelen tener longitudes diferentes, por lo que no se pueden convertir en tensores de tama√±o fijo. El relleno agrega un token de relleno especial para garantizar que las secuencias m√°s cortas tengan la misma longitud que la secuencia m√°s larga de un lote o la longitud m√°xima aceptada por el modelo. El truncamiento funciona en la direcci√≥n opuesta, truncando secuencias largas.

# truncation : El argumento de truncamiento controla el truncamiento. Puede ser un valor booleano o una cadena:

Verdadero o 'longest_first': trunca hasta una longitud m√°xima especificada por el argumento max_length o la longitud m√°xima aceptada por el modelo si no se proporciona max_length (max_length=None). Esto truncar√° token por token, eliminando un token de la secuencia m√°s larga del par hasta que se alcance la longitud adecuada.
'only_second': trunca hasta una longitud m√°xima especificada por el argumento max_length o la longitud m√°xima aceptada por el modelo si no se proporciona max_length (max_length=None). 'only_first': trunca hasta una longitud m√°xima especificada por el argumento max_length o la longitud m√°xima aceptada por el modelo si no se proporciona max_length (max_length=None). Esto solo truncar√° la primera oraci√≥n de un par si se proporciona un par de secuencias (o un lote de pares de secuencias).
Falso o 'do_not_truncate': no ‚Äã‚Äãse aplica ning√∫n truncamiento. Este es el comportamiento predeterminado.

# ¬øUn vector de alta dimensi√≥n?
El vector de salida del m√≥dulo Transformer suele ser grande. Generalmente tiene tres dimensiones:

Tama√±o del lote: la cantidad de secuencias procesadas a la vez (2 en nuestro ejemplo).
Longitud de la secuencia: la longitud de la representaci√≥n num√©rica de la secuencia (16 en nuestro ejemplo).

# **Relleno de las entradas**
La siguiente lista de listas no se puede convertir en un tensor:

batched_ids = [
    [200, 200, 200],
    [200, 200]
]

Para solucionar este problema, utilizaremos relleno para que nuestros tensores tengan una forma rectangular. El relleno garantiza que todas nuestras oraciones tengan la misma longitud agregando una palabra especial llamada token de relleno a las oraciones con menos valores. Por ejemplo, si tienes 10 oraciones con 10 palabras y 1 oraci√≥n con 20 palabras, el relleno garantizar√° que todas las oraciones tengan 20 palabras. En nuestro ejemplo, el tensor resultante se ve as√≠:

```python
padding_id = 100

batched_ids = [
    [200, 200, 200],
    [200, 200, padding_id],
]
```

El ID del token de relleno se puede encontrar en tokenizer.pad_token_id. Us√©moslo y enviemos nuestras dos oraciones a trav√©s del modelo de forma individual y en lotes:

```python
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)

sequence1_ids = [[200, 200, 200]]
sequence2_ids = [[200, 200]]
batched_ids = [
    [200, 200, 200],
    [200, 200, tokenizer.pad_token_id],
]

print(model(torch.tensor(sequence1_ids)).logits)
print(model(torch.tensor(sequence2_ids)).logits)
print(model(torch.tensor(batched_ids)).logits)
```

Hay algo mal con los logits en nuestras predicciones por lotes: la segunda fila deber√≠a ser la misma que los logits para la segunda oraci√≥n, ¬°pero tenemos valores completamente diferentes!

Esto se debe a que la caracter√≠stica clave de los modelos Transformer son las capas de atenci√≥n que contextualizan cada token. Estas tendr√°n en cuenta los tokens de relleno, ya que atienden a todos los tokens de una secuencia. Para obtener el mismo resultado al pasar oraciones individuales de diferentes longitudes a trav√©s del modelo o al pasar un lote con las mismas oraciones y el relleno aplicado, debemos indicarles a esas capas de atenci√≥n que ignoren los tokens de relleno.
