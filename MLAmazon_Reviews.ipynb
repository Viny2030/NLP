# ***¿Qué es CountVectorizer?***
CountVectorizer es una clase de scikit-learn que transforma una colección de documentos de texto en una matriz numérica de recuentos de palabras o tokens. Esta clase tiene una serie de parámetros que también pueden ayudar en tareas de preprocesamiento de texto, como la eliminación de palabras vacías, umbrales de recuento de palabras (es decir, máximos y mínimos), límites de vocabulario, creación de n-gramas y más. En este artículo, veremos cómo usar CountVectorizer de scikit-learn para preparar datos para su uso con un clasificador.

# **El clasificador Naive Bayes es un algoritmo de aprendizaje automático supervisado que se aplica comúnmente en casos de uso que involucran sistemas de recomendación, clasificación de texto y análisis de sentimientos. Debido a que funciona bien con conjuntos de datos con alta dimensionalidad, es un clasificador favorito para la clasificación de texto en particular.**

Naive Bayes (NB) también es un algoritmo de aprendizaje generativo, lo que significa que modela la distribución de puntos de datos para una clase o categoría determinada. Este clasificador probabilístico se basa en el teorema de Bayes, lo que significa que este clasificador bayesiano utiliza probabilidades condicionales y probabilidades previas para calcular las probabilidades posteriores.

Los clasificadores Naive Bayes funcionan de manera diferente, ya que operan bajo un par de supuestos clave, lo que le valió el título de ingenuo. Supone que los predictores en un modelo Naive Bayes son condicionalmente independientes o no están relacionados con ninguna de las otras características del modelo. También supone que todas las características contribuyen por igual al resultado.

Si bien estos supuestos se violan a menudo en situaciones del mundo real (por ejemplo, una palabra posterior en un correo electrónico depende de la palabra que la precede), simplifican un problema de clasificación al hacerlo más manejable computacionalmente. Es decir, ahora solo se requerirá una única probabilidad para cada variable, lo que, a su vez, hace que el cálculo del modelo sea más fácil. A pesar de este supuesto de independencia poco realista, el algoritmo de clasificación funciona bien, particularmente con tamaños de muestra pequeños.

# **Paso 6: Preprocesar los datos**
Después de dividir los datos, podemos comenzar a preprocesarlos. Esto incluye tareas de procesamiento de lenguaje natural, como tokenización, eliminación de palabras vacías, derivación y lematización.

Luego, usaremos una técnica popular de incrustación de palabras, llamada bag-of-words, para extraer características del texto. Esta técnica calcula específicamente la frecuencia de las palabras dentro de un documento determinado, lo que puede ayudarnos a clasificar documentos, suponiendo que documentos similares tienen contenido similar.

Podemos usar CountVectorizer o TfidfVectorizer de scikit-learn para que hagan el trabajo pesado por nosotros. Para los fines de este tutorial, solo mostraremos el código de TFidVectorizer, pero puede encontrar el código completo en el cuaderno en GitHub.


```python
def text_clean(text, method, rm_stop):
    text = re.sub(r"\n","",text)   #remove line breaks
    text = text.lower() #convert to lowercase
    text = re.sub(r"\d+","",text)   #remove digits and currencies
    text = re.sub(r'[\$\d+\d+\$]', "", text)
    text = re.sub(r'\d+[\.\/-]\d+[\.\/-]\d+', '', text)   #remove dates
    text = re.sub(r'\d+[\.\/-]\d+[\.\/-]\d+', '', text)
    text = re.sub(r'\d+[\.\/-]\d+[\.\/-]\d+', '', text)
    text = re.sub(r'[^\x00-\x7f]',r' ',text)   #remove non-ascii
    text = re.sub(r'[^\w\s]','',text)   #remove punctuation
    text = re.sub(r'https?:\/\/.*[\r\n]*', '', text)   #remove hyperlinks

    #remove stop words
    if rm_stop == True:
        filtered_tokens = [word for word in word_tokenize(text) if not word in set(stopwords.words('english'))]
        text = " ".join(filtered_tokens)

    #lemmatization: typically preferred over stemming
    if method == 'L':
        lemmer = WordNetLemmatizer()
        lemm_tokens = [lemmer.lemmatize(word) for word in word_tokenize(text)]
        return " ".join(lemm_tokens)

    #stemming
    if method == 'S':
        porter = PorterStemmer()
        stem_tokens = [porter.stem(word) for word in word_tokenize(text)]
        return " ".join(stem_tokens)

    return text
```
